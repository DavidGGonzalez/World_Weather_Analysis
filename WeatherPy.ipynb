### World Weather Analysis

```python
# Import Dependencies or Modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

### Zip() function
The `zip()` function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Create random latitudes and longitudes, then combine them
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
long = np.random.uniform(low=-180.000, high=180.000, size=1500)
lats_longs = zip(lats, long)
lats_longs
```

### List() function
The `list()` function returns a list in python

```python
# Now we add the lats_longs touples as a list of coordinates
coordinates = list(lats_longs)
coordinates
```

```python
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    country = citipy.nearest_city(coordinate[0],coordinate[1]).country_code
    print(f'City: {city}, Country Code: {country}')
```

```python
# Create a list to hold the cities
# Create a City's list
cities = []

# Get the nearest city using the coordinate and add it to the city's list if not found
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # if not found, then add it to the City's list
    if city not in cities:
        cities.append(city)
        
# Print total number of cities that were added by printing the lenght of the list
print(f'Total number of cities is: {len(cities)}')
```

```python
# Let's import the requests module so we can get data out of website via their API
import requests
import time
from datetime import datetime

# Get the API key from the config.py file
# ------ WE SHOULD NEVER PRINT THE KEY ------
from config import weather_key
```

```python
# let's create a variable to hold the URL we will use to get data using the WeatherMap API
url = "https://api.openweathermap.org/data/2.5/weather?units=imperial&appid=" + weather_key 
```

```python
# Create a list to hold city weather data
city_data = []

# Starting point
print("-----------------------------")
print("Start Data Retrieval         ")
print("-----------------------------")

# Ceate counters
# Due to the limitation imposed by the API package we got that only allows 60 calls per minute, we'll do it in batches of 50
record_count = 1
set_count = 1

# Loop through all the cities

# For testing purposes only
#for i in range(15):

for i in range(len(cities)):

    
    # Create sets of 50 cities
    if (i % 50 == 0 and i >= 50):
        # reset counter, increase set_counter, then set record_counter to 1
        set_count += 1
        record_count = 1
        
        # to bypass API limitations, set a timer to 60 seconds
        print("------------------------------------------------------------------")
        print(f"Waiting 60 seconds...")
        print("------------------------------------------------------------------")
        time.sleep(60)
        
    # Set city URL
    city_url = url + "&q=" + cities[i].replace(" ","+")
    print(f"Processing Record {record_count} of Set {set_count} | {cities[i]}")
    record_count += 1
    
    # Enclose the API request within a try-except so we can process the entire list of cities regradless of any error
    try:
        # Request the City weather via the API using the url and the request.get fucntion, in a json format
        city_weather = requests.get(city_url).json()
        
        # Parse data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        
        # Convert date/time using ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # Add city to the city_data list
        city_data.append({"City": cities[i],
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
    
    except:
        # If an error occurred, show it, city name should be above based on previous message
        # pass vs. continue
        # pass: does nothing, it will skip the to the next loop cycle
        # continue: will continue execution therefore would try to set the city_lat regardless if the API call failed or not, potentially generating a different error
        print(f"City was not found. Skipping...")
        pass
          
print("-----------------------------")
print("Data Retrieval Completed     ")
print("-----------------------------")

print(f"Total number of cities that were pulled via the API: {len(city_data)}")
```

```python
# Convert the array to a Pandas DataFrame
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Reorder columns in the DataFrame
new_order = ["City","Country","Date","Lat","Lng","Max Temp","Humidity","Cloudiness","Wind Speed"]
city_data_df = city_data_df[new_order]
city_data_df.head(10)
```

```python
# Export the data to a csv file
# Create output file
output_data_file = "weather_data/cities.csv"
# Export it
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
